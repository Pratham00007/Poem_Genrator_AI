```python
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential 
# memory of our model
from tensorflow.keras.layers import LSTM, Dense, Activation
from tensorflow.keras.optimizers import RMSprop
```

```python
#  shakespear poem text
# using directly online not dowloading it
filepath= tf.keras.utils.get_file('shakespeare.txt','https://storage.googleapis.com/download.tensorflow.org/data/shakespeare.txt')
```

```python
text=open(filepath,'rb').read().decode(encoding='utf-8').lower() 
# covering to all to lower to impove model accuracy as it predict next word
```

```python
text=text[300000:800000]
# trianing on the this part of neural network only to save time
```

```python

characters = sorted(list(set(text)))
# set to get unique and sorted to get sorted
```

```python
# CONVERT CHARACTER TO NUMERICAL

char_to_index = {char: i for i, char in enumerate(characters)}
# enumerate assign number to set values then we are storing that in dict as char as key and index as value
#  like {'f':1 , 'g':2}
```

```python

# AND AGAIN NUMERICAL TO CHARACTER

index_to_char = {i: char for i, char in enumerate(characters)}
# number as key and char as value
```

```python
# use 40 chracter to predict next character
seq_length=40
step_size=3
# how many chrac you want to shift to next sentence
```

```python
sentences=[]
next_characters=[]
```

```python
for i in range(0,len(text)- seq_length,step_size):
    sentences.append(text[i:i+seq_length])
    next_characters.append(text[i+seq_length])
```

* i jo hai wo 0 se length of text - sequence lengh tak jaega and step value step_size ke brabar
* -seq_length bcs jab last me phunge to sentence.append me index out of bound ho jaege
* next charcter next chracter lega

```python

x = np.zeros((len(sentences), seq_length, len(characters)), dtype=bool)
y = np.zeros((len(sentences), len(characters)), dtype=bool)
```

for x

| Argument          | Value / Meaning          | Description                                                                      |
| ----------------- | ------------------------ | -------------------------------------------------------------------------------- |
| `len(sentences)`  | Number of sequences      | Rows of x  total sequences we created from text                                 |
| `seq_length`      | 40 (example)             | Columns of x  length of each input sequence (number of characters per sequence) |
| `len(characters)` | Number of unique letters | Depth (3rd dimension)  one-hot encoding vector for each character               |
| `dtype=bool`      | True/False               | Each position will store 0/1 (False/True) instead of numbers                     |


for y

| Argument          | Value / Meaning          | Description                                                 |
| ----------------- | ------------------------ | ----------------------------------------------------------- |
| `len(sentences)`  | Number of sequences      | Rows of y  one target for each input sequence              |
| `len(characters)` | Number of unique letters | Columns  one-hot encoding for the **next character**       |
| `dtype=bool`      | True/False               | Each position will store 0/1 for the correct next character |



```python

for i, sentence in enumerate(sentences):
    for t, character in enumerate(sentence):
        x[i, t, char_to_index[character]] = 1
    y[i, char_to_index[next_characters[i]]] = 1
# in empty array just put one where the character occurs
print("x shape:", x.shape)
print("y shape:", y.shape)
```


### 1 `for i, sentence in enumerate(sentences):`

* `enumerate`  sequences ko **number** (i) ke saath iterate karta hai
* `sentence`  i-th sequence (40 characters ka string)
* Matlab: har sequence ko process karne ke liye loop

---

### 2 `for t, character in enumerate(sentence):`

* `t`  sequence ke character ka **position** (0 se 39)
* `character`  sequence ka t-th character
* Matlab: sequence ke **har character ko individually handle karna**

---

### 3 `x[i, t, char_to_index[character]] = 1`

* `i`  sequence number
* `t`  character position in sequence
* `char_to_index[character]`  character ka **one-hot index**
* `= 1`  mark kar diya ki **ye character is position par present hai**

 **Intuition:**
x me 3D array me har character ko 0/1 se represent kar rahe hain.

---

### 4 `y[i, char_to_index[next_characters[i]]] = 1`

* `i`  sequence number
* `next_characters[i]`  har sequence ke baad aane wala correct character
* `char_to_index[...]`  us character ka one-hot index
* `= 1`  mark kar diya ki ye character correct next character hai

 **Intuition:**
y me 2D array me har sequence ka **target next character** mark ho gaya.

---

### 5 `print("x shape:", x.shape)` / `print("y shape:", y.shape)`

* x.shape  `(num_sequences, seq_length, num_unique_characters)`
* y.shape  `(num_sequences, num_unique_characters)`
* Matlab: check kar rahe hain array **size aur dimensions** sahi hain ya nahi

---

###  Easy Summary (Hinglish)

1. Loop over each sequence  `i`
2. Loop over each character in sequence  `t`
3. x[i, t, char_index] = 1  **input sequence ko one-hot encode**
4. y[i, char_index_of_next_char] = 1  **next character ko one-hot encode**
5. Print shapes  confirm arrays are ready for training

---



```python
model=Sequential()
model.add(LSTM(128,input_shape=(seq_length,len(characters))))

# lstm= long short term memory
# retain prev input
# 128 neurons 
# input_shape=seq_length*len of chracters 
# get input and put them in lsdm layer

model.add(Dense(len(characters)))
# to add complexity , dense layer
model.add(Activation('softmax'))
# SCALES THE VALUE SO THAT THEY ADD UP TO ONE IE 100%

model.compile(loss='categorical_crossentropy',optimizer=RMSprop(learning_rate=0.01))
```

```python
model.fit(x,y,batch_size=256,epochs=4)
```

```python
model.save('textgenerator.h5')  # used .h5 instead of .model
```
